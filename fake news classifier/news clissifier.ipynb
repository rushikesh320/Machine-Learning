```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=data.dropna()
```

```python
messages=data.copy()
```

```python
messages.reset_index(inplace=True)##reset the index
```

```python
messages.head()
```

```python
import nltk
import re
```

```python
##data cleaning and processing remove stopwords and punctuation, split into word etc.
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review=review.lower()
    review=review.split()
    
    review=[ps.stem(word) for word in review if not word in stopwords.words('english') ]
    review=' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

```python
##preform bagging or countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,ngram_range=(1,3))
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
Y=messages['label']
```

```python
cv.get_feature_names()[:20]
```

```python
##split data into train and test data 
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(X,Y,test_size=0.33,random_state=0)
```

```python
countdf=pd.DataFrame(xtr,columns=cv.get_feature_names())
```

```python
countdf
```

```python
#train data to model
from sklearn.naive_bayes import MultinomialNB
clssi=MultinomialNB()
clssi.fit(xtr,ytr)
ypre=clssi.predict(xts)
```

```python
## check accuracy of model
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(yts,ypre)
```

```python
accuracy_score(yts,ypre)
```

passive agressive clissifier algorithm

```python
from sklearn.linear_model import PassiveAggressiveClassifier
licf=PassiveAggressiveClassifier(n_iter_no_change=50).fit(xtr,ytr)
ypre1=licf.predict(xts)
```

```python
confusion_matrix(yts,ypre1)
```

```python
accuracy_score(yts,ypre1)
```
